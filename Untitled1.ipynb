```python
x = cv2.imread("img1.png", 0)
y = cv2.imread("img1.png", 0)
z = cv2.imread("img2.png", 0)

def subdivide(volume, x, y, z):
    if(np.sum(x) * np.sum(y) * np.sum(z) == 0):
        volume.fill(0)
        return volume
    
    if len(x) <= 1:
        return volume

    m = len(volume) // 2
    
    volume[:m, :m, :m] = subdivide(volume[:m, :m, :m], x[:m, :m], y[:m, :m], z[:m, :m])
    volume[:m, :m, m:] = subdivide(volume[:m, :m, m:], x[:m, :m], y[:m, m:], z[:m, m:])
    volume[:m, m:, :m] = subdivide(volume[:m, m:, :m], x[:m, m:], y[:m, :m], z[m:, :m])
    volume[:m, m:, m:] = subdivide(volume[:m, m:, m:], x[:m, m:], y[:m, m:], z[m:, m:])
    volume[m:, :m, :m] = subdivide(volume[m:, :m, :m], x[m:, :m], y[m:, :m], z[:m, :m])
    volume[m:, :m, m:] = subdivide(volume[m:, :m, m:], x[m:, :m], y[m:, m:], z[:m, m:])
    volume[m:, m:, :m] = subdivide(volume[m:, m:, :m], x[m:, m:], y[m:, :m], z[m:, :m])
    volume[m:, m:, m:] = subdivide(volume[m:, m:, m:], x[m:, m:], y[m:, m:], z[m:, m:])
    
    return volume
```

```python
import cv2
import numpy as np
import time
from mpl_toolkits.mplot3d import Axes3D

x = cv2.imread("img1.png", 0)
y = cv2.imread("img1.png", 0)
z = cv2.imread("img2.png", 0)
d = len(x)
bounding_volume = np.ones((d,d,d), dtype=int)

start = time.time()
bounding_volume = subdivide(bounding_volume, x, y, z)
print(f"Recursive finished in {time.time() - start} seconds")

start = time.time()
for i in range(d):
    for j in range(d):
        for k in range(d):
            bounding_volume[i][j][k] = x[i][j] and y[i][k] and z[j][k]
print(f"Iterative finished in {time.time() - start} seconds")

            
colors = np.empty(bounding_volume.shape, dtype=object)
colors[bounding_volume.astype(bool)] = 'red'

start = time.time()
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.voxels(bounding_volume, facecolors=colors, edgecolor='k')
plt.show()
print(f"Drawing finished in {time.time() - start} seconds")
```
